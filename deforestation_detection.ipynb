```python
import numpy as np #loading numpy to load in the input image and the predicted mask
import matplotlib.pyplot as plt
```

```python
test = np.load('path to location of test image')
predict = np.load('path to location where predicted mask is stored')
```

```python
#defining the normalize function to visualize the input test image
def normalize(img):
    img = img.astype(np.float64)
    img -= np.min(img)
    img /= np.max(img)
    return img
```

```python
fig = plt.figure(figsize = (12,6))
figa = fig.add_subplot(1,2,1)
figa.imshow(normalize(test.transpose((1,2,0))),aspect='auto')
figa.set_title('Input RGB Composite')
figb = fig.add_subplot(1,2,2)
figb.imshow(predict[:,:],aspect='auto')
figb.set_title('Predicted Deforestation Mask')
```
